### üìù –ü—Ä–∞–∫—Ç–∏—á–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è:

1. –í–∏–≤–µ—Å—Ç–∏ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –≤–≤–µ–¥–µ–Ω–æ–≥–æ —á–∏—Å–ª–∞.
2. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤.
3. –°—Ç–≤–æ—Ä–∏—Ç–∏ —Ü–∏–∫–ª, —â–æ –≤–∏–≤–æ–¥–∏—Ç—å –¥–µ–Ω—å —ñ –Ω–æ–º–µ—Ä —Å–µ–∫—Ç–æ—Ä–∞ –∑ –Ω–æ–º–µ—Ä–æ–º —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É.




````python
# 1. –í–∏–≤–µ—Å—Ç–∏ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è –ø—Ä–æ —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –≤–≤–µ–¥–µ–Ω–æ–≥–æ —á–∏—Å–ª–∞.

# –ó–∞–ø–∏—Ç—É—î–º–æ —É –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –≤–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ –≤—Ç—Ä–∞—Ç
losses = int(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ (—á–∏—Å–ª–æ): "))

# –ü–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è —ñ –≤–∏–≤–æ–¥–∏–º–æ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–µ –ø–æ–≤—ñ–¥–æ–º–ª–µ–Ω–Ω—è
if losses == 0:
    print("–í—Ç—Ä–∞—Ç –Ω–µ–º–∞—î.")
elif 1 <= losses <= 10:
    print("–ú—ñ–Ω—ñ–º–∞–ª—å–Ω—ñ –≤—Ç—Ä–∞—Ç–∏.")
elif 11 <= losses <= 50:
    print("–°–µ—Ä–µ–¥–Ω—ñ –≤—Ç—Ä–∞—Ç–∏.")
else:
    print("–í–µ–ª–∏–∫—ñ –≤—Ç—Ä–∞—Ç–∏.")

# --------------------------------------------------------------

# 2. –ü–æ—Ä–∞—Ö—É–≤–∞—Ç–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤.

# –ó–∞–ø–∏—Ç—É—î–º–æ —É –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á–∞ –≤–≤–µ—Å—Ç–∏ —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É
unit1 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –ø–µ—Ä—à–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É (0-100): "))
unit2 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –¥—Ä—É–≥–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É (0-100): "))
unit3 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä–µ—Ç—å–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É (0-100): "))

# –û–±—á–∏—Å–ª—é—î–º–æ —Å–µ—Ä–µ–¥–Ω—î –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–Ω–µ
average_readiness = (unit1 + unit2 + unit3) / 3

# –í–∏–≤–æ–¥–∏–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∑ –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è–º –¥–æ –¥–≤–æ—Ö –∑–Ω–∞–∫—ñ–≤ –ø—ñ—Å–ª—è –∫–æ–º–∏
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤: {average_readiness:.2f}")

# --------------------------------------------------------------

# 3. –°—Ç–≤–æ—Ä–∏—Ç–∏ —Ü–∏–∫–ª, —â–æ –≤–∏–≤–æ–¥–∏—Ç—å –¥–µ–Ω—å —ñ –Ω–æ–º–µ—Ä —Å–µ–∫—Ç–æ—Ä–∞ –∑ –Ω–æ–º–µ—Ä–æ–º —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É.

# –ü—Ä–∏–ø—É—Å—Ç–∏–º–æ, —â–æ —É –Ω–∞—Å —î 5 –¥–Ω—ñ–≤ —ñ 3 —Å–µ–∫—Ç–æ—Ä–∏.
days = 5
sectors = 3

# –ó–æ–≤–Ω—ñ—à–Ω—ñ–π —Ü–∏–∫–ª - –ø–æ –¥–Ω—è—Ö
for day in range(1, days + 1):
    # –í–Ω—É—Ç—Ä—ñ—à–Ω—ñ–π —Ü–∏–∫–ª - –ø–æ —Å–µ–∫—Ç–æ—Ä–∞—Ö
    for sector in range(1, sectors + 1):
        # –û–±—á–∏—Å–ª—é—î–º–æ –Ω–æ–º–µ—Ä —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É (–º–æ–∂–Ω–∞ –ø—Ä–æ—Å—Ç–æ –ª—ñ—á–∏—Ç–∏ –ø–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä)
        calc_number = (day - 1) * sectors + sector
        # –í–∏–≤–æ–¥–∏–º–æ —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é
        print(f"–î–µ–Ω—å {day}, —Å–µ–∫—Ç–æ—Ä {sector}, –Ω–æ–º–µ—Ä —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É: {calc_number}")
````

### –ó–∞–≤–¥–∞–Ω–Ω—è 1. –°–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤. Gpt

````python
losses = int(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ (—á–∏—Å–ª–æ): "))
if losses == 0:
    print("–í—Ç—Ä–∞—Ç –Ω–µ–º–∞—î.")
elif 1 <= losses <= 10:
    print("–ú—ñ–Ω—ñ–º–∞–ª—å–Ω—ñ –≤—Ç—Ä–∞—Ç–∏.")
elif 11 <= losses <= 50:
    print("–°–µ—Ä–µ–¥–Ω—ñ –≤—Ç—Ä–∞—Ç–∏.")
else:
    print("–í–µ–ª–∏–∫—ñ –≤—Ç—Ä–∞—Ç–∏.")
````

### –ó–∞–≤–¥–∞–Ω–Ω—è 2. –†—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞ –Ω–∞ –æ—Å–Ω–æ–≤—ñ –≤–≤–µ–¥–µ–Ω–æ–≥–æ —á–∏—Å–ª–∞. Gpt

````python
unit1 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –ø–µ—Ä—à–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É (0-100): "))
unit2 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ –¥—Ä—É–≥–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É (0-100): "))
unit3 = float(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä–µ—Ç—å–æ–≥–æ –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—É (0-100): "))
average_readiness = (unit1 + unit2 + unit3) / 3
print(f"–°–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤: {average_readiness:.2f}")
````

### –ó–∞–≤–¥–∞–Ω–Ω—è 3. –¶–∏–∫–ª, —â–æ –≤–∏–≤–æ–¥–∏—Ç—å –¥–µ–Ω—å —ñ –Ω–æ–º–µ—Ä —Å–µ–∫—Ç–æ—Ä–∞ –∑ –Ω–æ–º–µ—Ä–æ–º —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É. Gpt

````python
days = 8
sectors = 9
for day in range(1, days + 1):
    for sector in range(1, sectors + 1):
        calc_number = (day - 1) * sectors + sector
        print(f"–î–µ–Ω—å {day}, —Å–µ–∫—Ç–æ—Ä {sector}, –Ω–æ–º–µ—Ä —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É: {calc_number}")
````

### –ó–∞–≤–¥–∞–Ω–Ω—è 1. –°–µ—Ä–µ–¥–Ω—ñ–π —Ä—ñ–≤–µ–Ω—å –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—ñ —Ç—Ä—å–æ—Ö –ø—ñ–¥—Ä–æ–∑–¥—ñ–ª—ñ–≤.

````python
Losses = int(input("–í–≤–µ–¥—ñ—Ç—å —Ä—ñ–≤–µ–Ω—å –≤—Ç—Ä–∞—Ç –ø—Ä–æ—Ç–∏–≤–Ω–∏–∫–∞(—á–º—Å–ª–æ):"))
if Losses == 0:
    print("–í—Ç—Ä–∞—Ç–∏ –Ω–µ –∑–Ω–∞—á–Ω—ñ")
elif 1<= Losses <= 20:
    print("–°–µ—Ä–µ–¥–Ω—ñ –≤—Ç—Ä–∞—Ç–∏")
elif 20<= Losses <= 70:
    print("–í–µ–ª–∏–∫—ñ –≤—Ç—Ä–∞—Ç–∏")
else:
    print("–ö–æ–ª–æ—Å–∞–ª—å–Ω—ñ –≤—Ç—Ä–∞—Ç–∏")
````
